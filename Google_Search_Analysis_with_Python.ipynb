```python
import pytrends
from pytrends.request import TrendReq
import pandas as pd
import matplotlib.pyplot as plt
```

```python
trends=TrendReq()
```

# **Now find the countries Where Certain Keywords are Searched The most**

```python
trends.build_payload(kw_list=['analytics'])
data=trends.interest_by_region()
data=data.sort_values(by='analytics',ascending=False)
data=data.head(20)
data
```

```python
data.reset_index().plot(x='geoName',y='analytics',kind='bar',figsize=(25,10),fontsize=20)
```

```python
data=TrendReq(hl='en-US',tz=360)
data.build_payload(kw_list=['data analysis'])
data=data.interest_over_time()
plt.subplots(figsize=(20,8))
data['data analysis'].plot()
plt.xlabel('Year')
plt.ylabel("total count")
plt.title('total google Searches for keywords:"machine learning',fontweight='bold')
```
